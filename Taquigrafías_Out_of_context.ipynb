```python
import os
import requests
from bs4 import BeautifulSoup

# URL base del sitio web
BASE_URL = 'https://www.scjn.gob.mx'

# Directorio de descarga
DOWNLOAD_DIR = '/content/drive/MyDrive/Taquigráficas/Taquis_pdf'

# Iterar por cada página
for page_num in range(0, 358):
    # URL de la página actual
    url = f'{BASE_URL}/multimedia/versiones-taquigraficas?fecha=All&field_vsts_instancia_target_id_1=All&page={page_num}'

    # Obtener el contenido HTML de la página
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encontrar todas las filas de la tabla en la página
    rows = soup.find_all('tr')

    # Iterar por cada fila y extraer los valores de las columnas requeridas
    for row in rows:
        fecha_cell = row.find('td', {'headers': 'view-field-vsts-fecha-table-column'})
        instancia_cell = row.find('td', {'headers': 'view-field-vsts-instancia-table-column'})
        
        # Si se encontraron las dos columnas, agregar sus valores a las listas
        if fecha_cell and instancia_cell:
            fecha = fecha_cell.text.strip()
            instancia = instancia_cell.text.strip()

            # Verificar si el archivo ya existe en el directorio de descargas
            pdf_filename = f"{fecha}_{instancia}.pdf"
            if os.path.exists(os.path.join(DOWNLOAD_DIR, pdf_filename)):
                print(f'El archivo {pdf_filename} ya existe en la carpeta de descargas')
                continue

            # Encontrar el enlace al archivo PDF y descargarlo en el directorio
            pdf_link = row.find('a', href=lambda href: href and href.endswith('.pdf'))
            if pdf_link is not None:
                pdf_url = f"{BASE_URL}{pdf_link['href']}"
                pdf_filename = f"{fecha}_{instancia}.pdf"
                
                # Crear el directorio si no existe
                if not os.path.exists(DOWNLOAD_DIR):
                    os.makedirs(DOWNLOAD_DIR)

                # Descargar el archivo en el directorio
                with open(os.path.join(DOWNLOAD_DIR, pdf_filename), 'wb') as f:
                    response = requests.get(pdf_url)
                    f.write(response.content)

                # Imprimir el nombre del archivo descargado
                print(f'Descargado {pdf_filename}')

    # Imprimir la cantidad de archivos PDF encontrados en la página actual
    print(f'Explorando página {page_num}')

```

Modificar los archivos pdf por archivos txt

```python
import os
import aspose.words as aw

# Directorio de entrada y salida
input_folder = '/content/drive/MyDrive/Taquigráficas/Taquis_pdf'
output_folder = '/content/drive/MyDrive/Taquigráficas/Taquis_txt'

# Lista para almacenar los nombres de los archivos ya convertidos
converted_files = []

# Verificar si el directorio de salida existe y crearlo si no
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
else:
    # Escanear el directorio de salida y agregar los nombres de los archivos sin extensión a la lista de archivos convertidos
    for filename in os.listdir(output_folder):
        if filename.endswith('.txt'):
            converted_files.append(os.path.splitext(filename)[0])

# Iterar sobre los archivos en el directorio de entrada
for filename in os.listdir(input_folder):
    try:
        # Verificar si el archivo es un PDF y no ha sido convertido previamente
        if filename.endswith('.pdf') and os.path.splitext(filename)[0] not in converted_files:
            # Crear la ruta completa de entrada y salida
            input_file = os.path.join(input_folder, filename)
            output_file = os.path.join(output_folder, os.path.splitext(filename)[0] + '.txt')

            # Abrir el archivo PDF y guardarlo como texto sin formato
            doc = aw.Document(input_file)
            doc.save(output_file, aw.SaveFormat.TEXT)

            # Agregar el nombre del archivo convertido a la lista
            converted_files.append(os.path.splitext(filename)[0])

            # Mostrar el mensaje de descarga
            print(f'Convertido {filename} a formato TXT')
    except Exception as e:
        print(f'Error al convertir {filename}: {str(e)}')
```

Formar un Dataframe con cuatro columnas a partir del nombre de archivo, modificar el formato de fecha y formar un archivo csv.

GPT.jpg

"En este código, se crea una lista llamada "datos" para almacenar los datos de cada párrafo que contiene una palabra clave. Dentro del bucle for que itera sobre los archivos de texto, se extraen la fecha y la instancia del nombre del archivo usando split() y se agregan a la lista "datos" junto con el nombre del archivo y el texto del párrafo. Luego, se crea el dataframe usando la lista "datos" y se guarda como un archivo CSV. Las columnas del dataframe se definen como 'Archivo', 'Fecha', 'Instancia' y 'Texto'."


```python
import os
import re
import pandas as pd
import aspose.words as aw
from datetime import datetime

# Carpeta que contiene los archivos de texto
carpeta = '/content/drive/MyDrive/Taquigráficas/Taquis_txt'

# Palabras clave a buscar
palabras_clave = ['alcohol','ay', 'baño', 'broma', 'burla', 'café', 'cafecito', 
                     'canción','carcajada','cerveza', 'chascarrillo','chela',
                     'chisme','chicharrón','demonio','desayuno', 'diablo', 
                     'dios','disculpa', 'divino','dormir','enfado', 'enojo',
                     'golosina','grosería', 'güey','leche','libro','mal',
                     'malinterprete','misa','molestia','ofenda','ofender', 'ofensa',
                     'ofendid','pasillo','película','risa', 'santo','soez', 'suciedad',
                     'sueño','tamal','té','tonto','tontería', 'vulgar', 'wey']

# Diccionario de meses
meses = {'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04', 'mayo': '05', 'junio': '06',
         'julio': '07', 'agosto': '08', 'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'}

# Lista para almacenar los datos
datos = []

# Iterar sobre todos los archivos de texto en la carpeta
for archivo in os.listdir(carpeta):
    if archivo.endswith('.txt'):
      
        # Extraer la fecha y la instancia del nombre del archivo
        nombre, extension = os.path.splitext(archivo)
        fecha, instancia = nombre.split('_')
        fecha = fecha.lower()

        # Convertir el nombre del mes en un número
        for mes, numero in meses.items():
            fecha = fecha.replace(mes, numero)

        # Eliminar las palabras "de" de la fecha
        fecha = re.sub(r'\bde\b', '', fecha)

        # Convertir la fecha a formato YYYY-MM-DD
        fecha_dt = datetime.strptime(fecha, '%d %m %Y')
        fecha_nueva_str = fecha_dt.strftime('%Y-%m-%d')

        # Obtener los segmentos de la fecha del nombre del archivo
        year = fecha.split()[2]
        month = fecha.split()[1]
        day = fecha.split()[0]

        # Abrir el archivo de texto con Aspose.Words
        doc = aw.Document(os.path.join(carpeta, archivo))

        # Iterar sobre los párrafos del documento
        for parrafo in doc.get_child_nodes(aw.NodeType.PARAGRAPH, True):
            texto_parrafo = parrafo.get_text().lower()

            # Buscar las palabras clave en el texto del párrafo
            for palabra in palabras_clave:
                if re.search(r'\b{}\b'.format(palabra), texto_parrafo):

                    # Agregar los datos a la lista
                    datos.append([archivo, year, month, day, instancia, palabra, texto_parrafo])

# Crear un DataFrame a partir de la lista

df = pd.DataFrame(datos, columns=['nombre_archivo', 'year', 'month', 'day', 'instancia', 'palabra', 'texto_parrafo'])

# Guardar el DataFrame como un archivo CSV
df.to_csv('/content/drive/MyDrive/Taquigráficas/CVSTaquis.csv') 
```

Agrupar celdas de fechas, instancias, palabras clave y página

```python
import pandas as pd

# Lee el archivo csv
df = pd.read_csv('/content/drive/MyDrive/Taquigráficas/CVSTaquis.csv')

# Agrupa las filas por fecha e instancia y concatena los valores de texto_parrafo
df_agrupado = df.groupby(['nombre_archivo','year', 'month', 'day', 'instancia', 'palabra'], as_index=False)['texto_parrafo', 'palabra'].agg({'texto_parrafo': ' '.join, 'palabra': ', '.join})

# Genera la columna Headline
df_agrupado['Headline'] = df_agrupado['instancia'] + ': ' +'" ' +df_agrupado['palabra'].str.split(',').apply(lambda x: x[0])+'"'

# Selecciona las columnas en el orden deseado
df_agrupado = df_agrupado[['nombre_archivo','day','month','year','instancia', 'palabra','Headline', 'texto_parrafo']]

# Guarda el dataframe resultante en un nuevo archivo csv
df_agrupado.to_csv('/content/drive/MyDrive/Taquigráficas/Taquis_agrupado.csv', index=False)

df_agrupado.to_excel('/content/drive/MyDrive/Taquigráficas/SCJN_OutOfContext.xlsx')

# Imprime el dataframe
print(df_agrupado)
```

Actualizar plantilla de TimelineJS con dataframe creado proeviamente

```python
import openpyxl

# abrir el archivo de Excel existente
workbook = openpyxl.load_workbook('/content/drive/MyDrive/Taquigráficas/SCJN_OutOfContext.xlsx')

# obtener la hoja deseada
sheet1 = workbook['Sheet1']

# inicializar listas para almacenar los valores de cada columna
year_vals = []
month_vals = []
day_vals = []
headline_vals = []
text_vals = []

# recorrer todas las filas de cada columna y copiar los valores
for row in sheet1.iter_rows(min_row=2, min_col=1, max_col=9):
    year_vals.append(row[4].value)
    month_vals.append(row[3].value)
    day_vals.append(row[2].value)
    headline_vals.append(row[7].value)
    text_vals.append(row[8].value)

# cerrar el archivo
workbook.close()

# abrir el archivo de Excel de destino
dest_workbook = openpyxl.load_workbook('/content/drive/MyDrive/Taquigráficas/TimelineJS_Template.xlsx')

# obtener la hoja deseada
dest_sheet = dest_workbook['od1']

# escribir valores en las celdas correspondientes
for i in range(len(year_vals)):
    dest_sheet.cell(row=i+3, column=1, value=year_vals[i])
    dest_sheet.cell(row=i+3, column=2, value=month_vals[i])
    dest_sheet.cell(row=i+3, column=3, value=day_vals[i])
    dest_sheet.cell(row=i+3, column=10, value=headline_vals[i])
    dest_sheet.cell(row=i+3, column=11, value=text_vals[i])
    dest_sheet.cell(row=i+3, column=12, value='')

# limpiar columnas y celdas especificadas
for i in range(4, 16):
    dest_sheet.cell(row=3, column=i, value='')
dest_sheet.cell(row=2, column=4, value='')
dest_sheet.cell(row=2, column=5, value='')
dest_sheet.cell(row=2, column=6, value='')
dest_sheet.cell(row=2, column=7, value='')
dest_sheet.cell(row=2, column=8, value='')
dest_sheet.cell(row=3, column=4, value='')

# guardar el archivo actualizado
dest_workbook.save('/content/drive/MyDrive/Taquigráficas/TimelineJS_Template.xlsx')

```
